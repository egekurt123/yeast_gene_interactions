## Used Genes from the Constanzo Paper

```python

df = pd.read_excel('data/Data File S5_SAFE analysis_Gene cluster identity and functional enrichments.xlsx', sheet_name='Global net. cluster gene list')

df
```

```python
ExE = pd.read_csv('extracted_data/ExE_imputed_5.csv', sep=',', index_col=0)
NxN = pd.read_csv('extracted_data/NxN_imputed_10.csv', sep=',', index_col=0)
ExN = pd.read_csv('extracted_data/ExN_filled_20.csv', sep=',', index_col=0)
```

```python
df["ORF"].to_list()
```

```python
genes = df["ORF"].to_list()
ExE = ExE.loc[ExE.index.isin(genes)]
ExN = ExN.loc[ExN.index.isin(genes)]
NxN = NxN.loc[NxN.index.isin(genes)]
```

```python
ExE
```

```python
ExN
```

```python
NxN
```

```python
ExE = ExE.reset_index()
ExN = ExN.reset_index()
NxN = NxN.reset_index()

GxG = ExE.merge(ExN, on='gene_id', how='outer').merge(NxN, on='gene_id', how='outer')
GxG = GxG.set_index('gene_id')

print(f"{GxG.isnull().sum().sum() / (GxG.shape[0] * GxG.shape[1]) * 100:.2f}% NaN values")
```

```python
GxG = GxG.fillna(GxG.mean()).fillna(0)
GxG.to_csv("extracted_data/paper_all_interactions.csv", index=True)
```
