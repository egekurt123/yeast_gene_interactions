```python
import matplotlib.pyplot as plt
import numpy as np

def plot_r2_comparison():
    # Data from your results
    datasets = ['ExE', 'NxN', 'ExN']
    models = ['Lasso', 'Ridge', 'Random Forest', 'Neural Network']
    
    # R values for each dataset and model
    r2_data = {
        'ExE': [0.3161, 0.3588, 0.2266, 0.2804],
        'NxN': [0.2143, 0.2101, 0.0391, 0.1728], 
        'ExN': [0.3075, 0.3317, 0.1956, 0.3090]
    }
    
    colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D']
    
    fig, ax = plt.subplots(1, 1, figsize=(12, 7))
    
    x = np.arange(len(datasets))
    width = 0.2
    
    for i, model in enumerate(models):
        model_scores = [r2_data[dataset][i] for dataset in datasets]
        bars = ax.bar(x + i*width, model_scores, width, label=model, color=colors[i], alpha=0.8)
        
        # Add value labels on bars
        for bar, score in zip(bars, model_scores):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.005,
                   f'{score:.3f}', ha='center', va='bottom', fontsize=9, fontweight='bold')
    
    ax.set_ylabel('R Score', fontsize=12, fontweight='bold')
    ax.set_xlabel('Dataset', fontsize=12, fontweight='bold')
    ax.set_title('R Scores Comparison Across Datasets', fontweight='bold', fontsize=14)
    ax.set_xticks(x + width * 1.5)
    ax.set_xticklabels(datasets, fontsize=11)
    ax.legend(fontsize=10)
    ax.grid(axis='y', alpha=0.3)
    ax.set_ylim(0, 0.4)
    
    plt.tight_layout()
    plt.show()

# Call the function
plot_r2_comparison()
```
