## Is it possible to symmetrically replace nan values, since the datasource is not completely symmetric?

```python
def symmetrically_replace_nan(df):
    for row_gene in df.index:
        for col_gene in df.columns:
            if row_gene in col_gene:
                continue  # skip diagonal
            if pd.isna(df.loc[row_gene, col_gene]):
                # find symmetric value
                matching_cols = [c for c in df.columns if row_gene in c]
                matching_rows = df.index[df.index.str.contains(col_gene)]

                if len(matching_cols) > 0 and len(matching_rows) > 0:
                    sym_val = df.loc[matching_rows[0], matching_cols[0]]
                    if not pd.isna(sym_val):
                        df.loc[row_gene, col_gene] = sym_val
```

```python
symmetrically_replace_nan(ExE)
```

```python
(ExE == symmetrically_replace_nan(ExE)).sum().sum()
```

```python
(ExN == symmetrically_replace_nan(ExN)).sum().sum()
```
